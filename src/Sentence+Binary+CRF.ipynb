```python
def gen_ngrams(data):
    allstrings = ""
    for i in range(len(data)):
        for j in range(i+1,len(data)+1):
            if len(data[i:j])<=5:
                allstrings =allstrings+" "+str(data[i:j])
    return allstrings
```
